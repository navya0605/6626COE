```python
import matplotlib.pyplot as plt
```

```python
#line-chart
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y,linestyle="solid",linewidth=2,marker='X',color='green')
plt.title("simple line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#task: createa line plot showing the growth of a savings bank over 6 months with following
#x-axis:months{1-6}
#y-axis:savings amount
#customizations:green dashed line with markers at each point,labels and a title
x=['Jan','Feb','Mar','Apr','May','June']
y=[1000,600,1200,1500,1350,1700]
plt.bar(x,y,color='pink')
plt.plot(x,y,linewidth=2,linestyle='dotted',color='green',marker='o')
plt.xlabel("first Half months")
plt.ylabel("savings growth")
plt.title("growth of savings account in XYZ Bank")
plt.show()
```

```python
#bar-graph
categories=['a','b','c','d','e']
values=[10,15,7,12,9]
plt.bar(categories,values,color=['lavender','pink','lightblue','lightyellow','lightgreen'],edgecolor=['purple','red','blue','yellow','green'],linewidth=2)
plt.title("basic bargraph")
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.show()
```

```python
#histograph
ages=[20,25,22,25,27,28,30,31,34,35,35,5,89,90,12,3,4,56,78,68,67,54,56,48,41,49,47]
b=[20,30,40,50,60,70]
plt.hist(ages,bins=b,color='skyblue',edgecolor='black',linewidth=4)
plt.title("age distribution")
plt.xlabel("age ranges")
plt.ylabel("frequency")
plt.grid(True)
plt.show()
```

```python
#pie chart

lan = ['python', 'ruby', 'c++', 'java']
per = [40, 10, 20, 30]
plt.pie(per, labels=lan, colors=['violet', 'hotpink', 'lightgreen', 'aqua'],autopct='%1.1f%%',startangle=180)
plt.title("Programming Languages Percentage")
plt.legend()
plt.show()
```

```python
#scatter plot
x=[5,10,15,20,25]
y=[10,20,25,30,35]
plt.scatter(x,y,color='magenta',s=100,edgecolor='black',marker='*')
plt.title("scatter plot example")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```
